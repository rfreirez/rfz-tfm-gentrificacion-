# Análisis de evolución de factores de gentrificación

### <span style="color:red">1. Indicador Edad</span>

#### <span style="color:blue">1.1 Distribución de la variable Nro_Personas entre 15 a 44 años</span>

```python
# Lectura csv edad
dfEdad = pd.read_csv('../../data_preprocesada_N3/clean_edad_barrios_menores65.csv')

# Plot evolución de numero de habitantes en barrios por anios
g = sns.FacetGrid(dfEdad, col="BARRIOS", col_wrap=5, height=4, margin_titles=True)
g.map(sns.barplot, "ANIO", "NRO_PERSONAS")
g.set_axis_labels("Anios", "Nro.Personas")
g.set_titles(row_template = '{row_name}', col_template = '{col_name}')
```
